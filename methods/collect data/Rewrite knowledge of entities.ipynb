```python
data_path = '2022-GLOBAL-AI-CHALLENGE'
```

```python
import json
with open(f'{data_path}/kg.json', encoding='utf-8') as f:
    data = json.load(f)

print(type(data))
```

```python
data = list(data.items())
```

Concatenate attributes of each entity as a single sentence.

```python
import pandas as pd
```

```python
kg_data = pd.DataFrame(columns=['entity','knowledge'])
kg_data
```

```python
from tqdm import tqdm

for j in tqdm(range(len(data))) :
    i = data[j]
    entity = i[0]
    attrs = i[1]
    knowledge = ''
    for m in attrs:
        knowledge += f'[ett] 的{m[1]}是{m[2]}。'
    knowledge= knowledge[:-2]

    kg_data.loc[j,'entity'] = entity
    kg_data.loc[j,'knowledge'] = knowledge
    
```

```python
kg_data['domain'] = 0
kg_data
```

```python
klg_data = kg_data
length = len(klg_data)
```

```python
for i in range(length):
    content = str(klg_data.loc[i,'knowledge'])
    if '国籍' in content or '出生地' in content or '出生时间' in content or 'Birth place' in content or '出生日期' in content or '籍贯' in content or '生日' in content or '作者简介' in content: klg_data.loc[i,'domain'] = 'figure'
    if '势力' in content or '亲属' in content or '部落' in content or '类型' in content: klg_data.loc[i,'domain'] = 'game figure'
    if '近义词' in content or '反义词' in content or '释义' in content or '拼音' in content: klg_data.loc[i,'domain'] = 'idiom'
    if '诗句' in content or '诗词全文' in content or '名句' in content : klg_data.loc[i,'domain'] = 'poetry'
    if '出版社' in content or '出版时间' in content or '价格' in content: klg_data.loc[i,'domain'] = 'literature'
    if '主演' in content or '上映时间' in content or '导演' in content: klg_data.loc[i,'domain'] = 'film and television'
    if '周边景点' in content or '开放时间' in content or '门票' in content: klg_data.loc[i,'domain'] = 'tourist attractions'
    if '辅料' in content or '口味' in content: klg_data.loc[i,'domain'] = 'food'
    if '谱曲' in content or '专辑' in content or (('歌曲' in content) and ('音乐' in content)) : klg_data.loc[i,'domain'] = 'music'
    if '动物' in content or '植物' in content:  klg_data.loc[i,'domain'] = 'nature'
    
    if klg_data.loc[i,'domain'] == 0: klg_data.loc[i,'domain'] = 'others'
```
